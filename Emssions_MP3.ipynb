```python
import matplotlib.pyplot as plt

m_h2_cyc = 6.260247*10**-7 #kg/cycle
m_c8h18_cyc = 1.5657265*10**-6 #kg/cycle

m_h2 = 4.032 #kg/kmol
m_h2o = 18.01528 #kg/kmol
m_c8h18 = 114.23 #kg/kmol
m_co2 = 44.01 #kg/kmol

rpm = 3000 #cycles/min

mratio_h2_h2o = m_h2o/(2*m_h2) #kg h2o per kg h2
mratio_c8h18_co2 = 8*m_co2/m_c8h18 #kg co2 per kg c8h18

def yearly_emissions(mpc,mass_ratio): #calculates yearly emissions from known mass intake per cycle and usage over a year
    ans = mpc*mass_ratio*rpm*30*365*0.7 #assuming the engine is operated 30min/day, 70% of the year
    return ans #kg of emissions

Emissions_H2O = yearly_emissions(m_h2_cyc, mratio_h2_h2o)
Emissions_CO2 = yearly_emissions(m_c8h18_cyc, mratio_c8h18_co2)

Emissions_H2O, Emissions_CO2
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
products = ['H2O', 'CO2']
masses = [Emissions_H2O, Emissions_CO2]

ax.bar(products, masses, edgecolor='black', hatch=['.', '//'])
ax.set_ylabel('Kg')
ax.set_title('Yearly Emissions Comparison of Iso-Octane and Hydrogen')


ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.rcParams.update({'font.size': 20})

plt.show()
```
